```python
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski
from skimage import data, color
from skimage.filters import threshold_otsu, threshold_local, try_all_threshold
```

## Show/Load Image Functions


```python
def show_image(image, title="Image", cmap_type="gray"):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis("off")
    plt.show()


def load_image(path):
    image = plt.imread(path)
    return image
```

```python
coffee = data.coffee()
print(coffee.shape)
show_image(coffee)

coins = data.coins()
print(coins.shape)
show_image(coins)
```

## RGB to GrayScale


```python
# Load the rocket image
rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)

# Show the original image
print(rocket.shape)
show_image(rocket, "Original RGB image")

# Show the grayscale image
print(gray_scaled_rocket.shape)
show_image(gray_scaled_rocket, "Grayscale image")
```

## Numpy on Images

### Flipping Image


```python
seville_vertical_flip = load_image("datasets/1/sevilleup(2).jpg")
show_image(seville_vertical_flip, "Seville vertical flip")

seville_horizontal_flip = np.flipud(seville_vertical_flip)
show_image(seville_horizontal_flip, "Seville horizontal flip")

seville = np.fliplr(seville_horizontal_flip)
show_image(seville, "Seville")
```

### Histograms


```python
image = load_image("datasets/1/4.1.01.tiff")

# Obtain the red channel
red_channel = image[:, :, 0]

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# Set title and show
plt.title("Red Histogram")
plt.show()
```

## Threshold
### Global Thresholding
#### Global: Uniform lighting and background
#### Local: Variation in lighting

```python
chess_pieces_image = load_image("datasets/1/bw.jpg")
show_image(chess_pieces_image, "Chess pieces")

chess_pieces_image_gray = color.rgb2gray(chess_pieces_image)
show_image(chess_pieces_image_gray, "Chess pieces gray")

# Obtain the optimal threshold value with otsu
thresh = threshold_otsu(chess_pieces_image_gray)

# Apply thresholding to the image
binary = chess_pieces_image_gray > thresh

# Show the image
show_image(binary, 'Binary image')
```

### Different Thresholds

```python
fruits = load_image("datasets/1/fruits-2.jpg")
show_image(fruits, "Fruits")

fruits_gray = color.rgb2gray(fruits)
show_image(fruits_gray, "Fruits gray")

thresh = try_all_threshold(fruits_gray, verbose=False)
```
